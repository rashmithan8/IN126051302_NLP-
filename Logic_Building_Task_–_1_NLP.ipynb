1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are treated as different tokens if case normalization is not applied.
"Love" may appear at the beginning of a sentence, while "love" appears elsewhere.
Converting to lowercase ensures both are treated as the same word → improves consistency.

2. What happens if stopwords are not removed?

Increases noise in data
Adds unnecessary words like "is", "the", "and"
Reduces model performance
Increases computation time

3. When removing stopwords can be harmful

Sentiment Analysis
Example: "I am not happy" → removing "not" changes meaning
Question Answering Systems
Example: "What is AI?" → removing "what" breaks the query

4.      Stemming vs Lemmatization

Feature	Stemming	Lemmatization

Method	Cuts words	Uses vocabulary

Output	Not always meaningful	Always meaningful

Example	"running" → "run" or "runn"	"running" → "run"

Accuracy	Low	High

```python
#TASK 2
import re
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    text = re.sub(r"\d+", "", text)
    text = text.lower()
    text = re.sub(r"(.)\1{2,}", r"\1", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    tokens = text.split()
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]
    clean_sentence = " ".join(tokens)
    return tokens, clean_sentence
```

```python
#TASK 3
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product",
    "Worst service ever... 0/10",
    "Call me at 9999999990",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
results = []
for text in test_sentences:
    tokens, clean = preprocess_text(text)
    results.append((text, tokens, clean))
for r in results:
    print("\nOriginal:", r[0])
    print("Tokens:", r[1])
    print("Clean:", r[2])
```

```python
#TASK 4
def analyze(tokens):
    if not tokens:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return total, unique, round(avg_len, 2)

for text, tokens, clean in results:
    total, unique, avg = analyze(tokens)
    print(f"\nText: {text}")
    print(f"Total Tokens: {total}")
    print(f"Unique Tokens: {unique}")
    print(f"Avg Length: {avg}")
```

```python
#TASK 5
from collections import Counter

all_tokens = []
for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
#TASK 6
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
